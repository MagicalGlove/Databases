```python
def insert_doc(document):
    try:
        collection.insert_one(document)
        print("Document inserted successfully.")
    except Exception as e:
        print(f"Error inserting document: {e}")
```

```python
for index, row in df.iterrows():
    doc = { "city_id": row["Account Number"],
            "reporting_year": row["Reporting Year"],
            "measurement_year":row["Measurement Year"],
            "boundary": row["Boundary"],
            "primary_methodology":row["Primary Methodology"],
            "methodology_details":row["Methodology Details"],
            "gases": row["Gases included"].split(";") if pd.notna(row["Gases included"]) else [],
            "total_emissions": row["Total City-wide Emissions (metric tonnes CO2e)"],
            "total_scope_1": row["Total Scope 1 Emissions (metric tonnes CO2e)"],
            "total_scope_2": row["Total Scope 2 Emissions (metric tonnes CO2e)"],
            "increase_decrease": row["Increase/Decrease from last year"],
            "reason_for_increase":row["Reason for increase/decrease in emissions"]
            }
    insert_doc(doc)
```
